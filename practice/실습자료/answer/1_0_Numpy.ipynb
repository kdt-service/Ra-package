## Numpy의 필요성

```python
## 행렬합 구하기
```

```python
matrix_1 = [[1, 2], [3, 4]]
matrix_2 = [[5, 6], [7, 8]]
```

```python
matrix_1 + matrix_2
```

```python
new_matrix = []
for i, row in enumerate(matrix_1):
    # print(row)
    temp = [] # 새로운 행을 위한 임시 리스트
    for j, value in enumerate(row):
        # print(value)
        # matrix_1, matrix_2의 i, j 번째 위치의 값을 더함
        sum_value = matrix_1[i][j] + matrix_2[i][j] 
        temp.append(sum_value)
    new_matrix.append(temp)
new_matrix
```

```python
# temp라는 행을 만들어주는 임시 리스트가 없다면?
# [6, 8, 10, 12]
```

```python
# numpy ndarray 객체로 변환
npm_1 = np.array(matrix_1)
npm_2 = np.array(matrix_2)
```

```python
print(npm_1)
print(npm_2)
```

```python
npm_1 + npm_2
```

```python
## 모든 요소에 1씩 더하기
```

```python
new_matrix = []
for i, row in enumerate(matrix_1):
    temp = []
    for j, value in enumerate(row):
        # print(value + 1)
        temp.append(value + 1)
    new_matrix.append(temp)
new_matrix
```

```python
npm_1 + 1 # 2차원, 0차원 더하기 -> 브로드캐스트
```

## Numpy vs Python 속도 비교

```python
import time
size = 10000000
# list
start_time = time.time()
x = list(range(size))
y = list(range(size))
z = [x[i] + y[i] for i in range(size)]
print("리스트 걸린시간", time.time() - start_time)

# adlist
start_time = time.time()
x = np.arange(size)
y = np.arange(size)
z = x + y
print("넘파이 걸린시간", time.time() - start_time)
```

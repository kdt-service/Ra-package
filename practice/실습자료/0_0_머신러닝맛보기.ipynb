### 예시) 보스턴 집 값 예측

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
data_boston = load_boston()
data_boston
```

```python
print(data_boston.DESCR)
```

```python
data_boston.keys()
```

```python
# input value : data_boston.data
# target value : data_boston.target
data_boston.data.shape
```

```python
data_boston.target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_boston.data, data_boston.target)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 1번. 모델 호출
model = LinearRegression()
model
```

```python
# 2번. 모델 학습
model.fit(data_boston.data, data_boston.target)
```

```python
# 3번. 학습결과 확인
model.coef_ 
```

```python
model.intercept_
```

```python
# 4번. 모델 검증
model.score(data_boston.data, data_boston.target)
```
